```python
import numpy as np
t = [i*0.1 for i in range(20)] # Wall Thickness [mm]
D = [i*0.2 for i in range(20)] # Diameter [mm]
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
t = np.linspace(0.1, 1.0, n)/1000.0
d = np.linspace(10.0,30.0, n)/1000.0

T, D  = np.meshgrid(t, d)
T_plt, D_plt  = np.meshgrid(t*1000.0, d*1000.0)
sigma = 2.0 * 1.0e+6 #[Pa]

# sigma = (P*D)/(2*T)
# P = sigma * (2*T)/D

P  = sigma * (2*T)/D #[Pa]
P  = P/133.3223684   #[Pa]->[mmHg]

lvls = np.linspace(0,3200,20)

cf = plt.contourf(T_plt, D_plt, P, levels=lvls)
cp = plt.contour(T_plt, D_plt, P, levels=lvls, colors='k')

plt.clabel(cp, lvls,  # label every second level
           inline=1,
           fmt='%1.1f',
           fontsize=8)

plt.xlabel('Wall Thickness of Blood Vessel[mm]')
plt.ylabel('Diameter of Blood Vessel[mm]')
plt.title('Burst Blood Pressure [mmHg] '+'for SigmaMax[MPa]='+str(sigma/1.0e+6))

plt.colorbar(cf)
plt.show()
```
